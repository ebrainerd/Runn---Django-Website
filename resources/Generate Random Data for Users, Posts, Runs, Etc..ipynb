## Generate Users

10% of users are coaches. 30% of users belong to a team. Coaches cannot belong to a team. 

```python
users = pd.DataFrame(columns=["UserID", "FirstName", "LastName", "Email", "Tagline", "Location", "UserType", "Team"])
users.head()
```

```python
userID = 0
for i in range(100):
    user = {}
    user["UserID"] = userID
    userID += 1
    sample_user = fn.sample(1)
    user["FirstName"] = sample_user["GivenName"].values[0]
    user["LastName"] = sample_user["Surname"].values[0]
    user["Email"] = sample_user["EmailAddress"].values[0]
    user["Tagline"] = random.choice(flat_lorem)
    user["Location"] = sample_user["City"].values[0] + ", " + sample_user["State"].values[0]
    
    if np.random.random() < 0.1:
        user["UserType"] = 1 # 1 means user is a coach
    else:
        user["UserType"] = 0 # 0 means user is an athlete 
        
    users = users.append(user, ignore_index = True)
    
# assign 30% of users to a coach (if they are not a coach themselves)
coach_ids = users[users["UserType"] == 1]["UserID"].values
user_team = []
for i in range(len(users)):
    if users.loc[i]["UserType"] == 1:
        user_team.append(np.nan)
        continue
    if np.random.random() < 0.3: # 30% of users are on a team 
        user_team.append(np.random.choice(coach_ids))
    else:
        user_team.append(np.nan)
users["Team"] = pd.Series(user_team, index=users.index, dtype="Int64")
```

## Generate Runs

Will generate 1000 random runs. 5 users will not have any runs. Other 95 users will have at least 1 run. 50% of runs will have an image posted with their run. 80% of users will post some text about their run. 

```python
# generates a random date between a specified start and end date 
# https://stackoverflow.com/questions/553303/generate-a-random-date-between-two-other-dates
start_date = datetime.datetime(2018, 1, 1, 0, 0, 0)
end_date = datetime.datetime(2020, 2, 28, 0, 0, 0)
def random_date(start, end):
    """
    This function will return a random datetime between two datetime 
    objects.
    """
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = random.randrange(int_delta)
    return start + datetime.timedelta(seconds=random_second)
```

```python
runs = pd.DataFrame(columns=["RunID", "UserID", "Distance", "RunTime", "Date", "Location", "Image", "Post", "Likes"])
runs.head()
```

```python
userID_list = users["UserID"].to_list()
users_no_runs = np.random.choice(userID_list, size=5, replace=False)
runID = 0
img = 0

while (len(runs) < 1000):
    if len(userID_list) > 0:
        temp_userID = random.choice(userID_list)
        userID_list.remove(temp_userID)
    else:
        temp_userID = users["UserID"].sample(1).values[0]
        
    if temp_userID in users_no_runs:
        continue
        
    run = {}
    run["RunID"] = runID
    runID += 1
    run["UserID"] = temp_userID
    # uniform random distances between 0.5 mile and 15 miles (rounded to 1 decimal place)
    run["Distance"] = round(np.random.uniform(low=0.5, high=15), 1) 
    # running time is distance * pace (where pace is in seconds per mile)
    run_time_seconds = round(run["Distance"] * np.random.randint(low=6*60, high=13*60), 0) # seconds per mile
    run["RunTime"] = str(datetime.timedelta(seconds=run_time_seconds)) # convert time to HH:MM:SS
    run["Date"] = random_date(start_date, end_date) #.strftime("%Y/%m/%d at %H:%M:%S")
    city_state = fn[["City","State"]].sample(1)
    run["Location"] = city_state["City"].values[0] + ", " + city_state["State"].values[0]
    
    if np.random.random() < 0.5:
        run["Image"] = "IMG" + f"{img:04}"
        img += 1
    else:
        run["Image"] = np.nan
    
    if np.random.random() < 0.8:    
        # randomly choose how many sentences in the post (between 1 and 5)
        num_sentences = np.random.randint(1, 5+1)
        run["Post"] = " ".join(np.random.choice(flat_lorem, size=num_sentences))
    else:
        run["Post"] = np.nan
        
    run["Likes"] = 0
        
    runs = runs.append(run, ignore_index = True)
```

## Generate Comments

5 users will not have any comments. 70% of runs will have comments. Of those runs with comments, there will be between 1 and 5 comments. 

```python
comments = pd.DataFrame(columns=["CommentID","UserID","RunID","CommentText","Time"])
comments.head()
```

```python
userID_list = users["UserID"].to_list()
users_allowed_comments = np.random.choice(userID_list, size=95, replace=False)
# copy of the above list because of the users that are allowed to comment, we want to make sure 
# each comments at least once. So we remove their ID from the copied list after commenting.
# When there are no more user IDs in the copied list, we sample from the original list
users_allowed_comments_copy = list(np.copy(users_allowed_comments))
commentID = 0

for run_id in runs.index:
    # 70% of runs have a comment
    if np.random.random() < 0.7:
        # randomly choose between 1 and 5 comments per run 
        for i in range(np.random.randint(1,5+1)):
            comment = {}
            comment["CommentID"] = commentID
            commentID += 1
            if len(users_allowed_comments_copy) > 0:
                userID = np.random.choice(users_allowed_comments_copy)
                users_allowed_comments_copy.remove(userID)
            else:
                userID = np.random.choice(users_allowed_comments)
            comment["UserID"] = userID
            comment["RunID"] = run_id
            # randomly choose how many sentences in the comment (between 1 and 5)
            num_sentences = np.random.randint(1, 5+1)
            comment["CommentText"] = " ".join(np.random.choice(flat_lorem, size=num_sentences))
            
            # comment time must come after the run time
            start_date = runs.loc[run_id]["Date"]
            comment["Time"] = random_date(start_date, end_date) #.strftime("%Y/%m/%d at %H:%M:%S")
            comments = comments.append(comment, ignore_index = True)
```

```python
comments.to_csv("comments.csv", index=False, header=True)
```

## Generate Following

5 users will not be following any other users. Of those who follow other users, they will follow between 1 and 80 users. Users cannot follow themselves.

```python
following = pd.DataFrame(columns=["UserID_1", "UserID_2"])
following
```

```python
userID_list = users["UserID"].to_list()
users_allowed_following = np.random.choice(userID_list, size=95, replace=False)

for user in users_allowed_following:
    # remove user from userID_list because a user cannot follow themselves
    userID_list_copy = list(np.copy(userID_list))
    userID_list_copy.remove(user)
    for i in range(np.random.randint(1,80+1)):
        f = {}
        f["UserID_1"] = user
        f["UserID_2"] = np.random.choice(userID_list_copy)
        userID_list_copy.remove(f["UserID_2"])
        following = following.append(f, ignore_index = True)
    
```
